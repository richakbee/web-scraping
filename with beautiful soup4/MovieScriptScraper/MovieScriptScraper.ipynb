```python
mainurl="https://www.imsdb.com"
```

## Get A Page

```python
def getPage(url):
        source=requests.get(url).text
        soup=BeautifulSoup(source,'html5lib')
        return soup
```

```python
article=getPage(mainurl)
print(article.prettify)
```

```python
article.find_all('a',href=re.compile('/genre'))
```

### get the genres

```python
genre=[]
for genres in article.find_all('a',href=re.compile('/genre')):
    genre.append(genres.text)
genre    
```

### get main page for a genre

```python
if user_genre in genre:
     soup=getPage(mainurl+'/genre/'+user_genre)
else:
    print("Unknown genre")
          
```

```python
print(soup.prettify)
```

### get Titles and IMSRB_title URL's

```python
title_url=soup.find_all('a',href=re.compile('/Movie Scripts'))[5:]
```

```python
Title_name=[]
IMSRB_title_url=[]
for title_urls in title_url:
    Title_name.append(title_urls.text)
    IMSRB_title_url.append(mainurl+str(title_urls.get('href')))
```

```python
 Title_name
```

```python
a=getPage(IMSRB_title_url[0])
```

```python
mainurl+a.find('a',href=re.compile('/scripts')).get('href')
```

```python
len(IMSRB_title_url)
```

### get IMSRB scripts URL's

```python
IMSRB_scripts_url=[]
for scripts in IMSRB_title_url:
        script_pages=getPage(scripts)
        #to handle cases where the scripts links are missing
        if(type(script_pages.find('a',href=re.compile('/scripts'))) !=type(None)):
            IMSRB_scripts_url.append(mainurl+script_pages.find('a',href=re.compile('/scripts')).get('href'))
        else:
            IMSRB_scripts_url.append('NA')
```

```python
IMSRB_scripts_url[:5]
```

```python
len(IMSRB_scripts_url)
```

## Save Data as Data Frame

```python
movie_dataframe=pd.DataFrame({'Title': Title_name ,'IMSRB_title_url' :IMSRB_title_url,'IMSRB_scripts_url':IMSRB_scripts_url})
```

```python
movie_dataframe.head()
```

```python
movie_dataframe.to_csv(user_genre+'GenreScript.csv',index=False)
```
