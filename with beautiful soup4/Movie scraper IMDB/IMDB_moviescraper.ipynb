```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

## Get URL's of all pages

```python
df=title_codes.copy()
df['url']=['https://www.imdb.com/title/'+x+'/?ref_=adv_li_tt' for x in df['codes'].values]
```

```python
df.head(11)
```

## Get content from a page

```python
source=requests.get(df['url'][0]).text
soup=BeautifulSoup(source,'lxml')
```

```python
print(soup.prettify())
```

### Get and parse a page

```python
def getpage(url):
    source=requests.get(url).text
    soup=BeautifulSoup(source,'lxml')
    return(soup)
```

### Finding all headers

<img src="img/headers.png">

```python
for val in soup.find_all('h4',class_='inline'):
    print(val.text.split(':')[0])
```

```python
#kkeping only required headers
val_required=['Director','Writer','Star','Genre','Language','Release Date','Also Known As','Budget','Opening Weekend USA',
              'Gross USA','Cumulative Worldwide Gross','Runtime']
```

```python
#getting data only for required headers
details=soup.find_all('h4',class_='inline')
for val in details:
    if (val.text.split(':')[0]) in val_required:
        print(val.findParent('div'))
        print()
```

### Scrape for single movie

<b>For some movies have multiple directors / writers
<img src="img/directors.png">

<b>And some have Single Director or Writer
<img src="img/director.png">

```python
details=soup.find_all('h4',class_='inline')
for val in details:
    v=val.text.split(':')[0]
    if len(re.compile('|'.join(val_required)).findall(v))>0:
        if len(re.compile("Director").findall(v))>0:
            print([x.text for x in val.findParent('div').find_all('a')])
        if len(re.compile("Writer").findall(v))>0:
            print([x.text for x in val.findParent('div').find_all('a')])
        if v=="Stars":
             print([x.text for x in val.findParent('div').find_all('a')])
        if v=="Genres":
            print([x.text for x in val.findParent('div').find_all('a')])
        if v=="Language":
            print([x.text for x in val.findParent('div').find_all('a')])
        if v=="Release Date":
            print(' '.join(val.next_sibling.strip().split(' ')[0:3]))
        if v=="Also Known As":
            print(val.next_sibling.strip())
        if v=="Budget":
            print(val.next_sibling.strip())
        if v=="Opening Weekend USA":
            print(val.next_sibling.strip())
        if v=="Gross USA":
            print(val.next_sibling.strip())
        if v=="Cumulative Worldwide Gross":
            print(val.next_sibling.strip())
        if v=="Runtime":   
            print(val.findParent('div').find('time').text.strip())
         
```

## Scrape all movies data

### initalize the data frame to store data

```python
#fetching it for all the movies
import numpy as np
import json
df['Directors']=np.NAN
df['Writers']=np.NAN
df['Stars']=np.NAN
df['Genres']=np.NAN
df['Language']=np.NAN
df['Release_Date']=np.NAN
df['Title']=np.NAN
df['description']=np.NAN
df['Budget']=np.NAN
df['opening_Weekend_USA']=np.NAN
df['Gross_USA']=np.NAN
df['Cumm_world_gross']=np.NAN
df['runtime']=np.NAN
df['MPAA rating']=np.NAN

for page_url_num in range(len(df.url)):
    url=df.url[page_url_num]
    soup=getpage(url)
    
    json_data = soup.find('script', type="application/ld+json")
    data=json.loads(json_data.text)
    
    
    
    details=soup.find_all('h4',class_='inline')
    i=page_url_num
    
    df['Title'][i]=data["name"]
    df['description'][i]=data["description"]
    df['MPAA rating'][i]=data["contentRating"]
    df['Genres'][i]=data["genre"]
    
    if type(data["director"])==dict:
        director=[data["director"]]
    else: 
        director=data["director"] 
    df['Directors'][i]=[directors.get('name') for directors in director]
    
    if type(data["creator"])==dict:
        creator=[data["creator"]]
    else:
        creator=data["creator"]
    temp=[]    
    for dicts in creator:
        for keys in dicts:
            if keys=='name':
                 temp.append(dicts.get('name')) 
            else:
                pass        
    df['Writers'][i]=temp
    
       
    
    if type(data["actor"])==dict:
        actor=[data["actor"]]
    else: 
        actor=data["actor"]    
    df['Stars'][i]=[actors.get('name') for actors in actor]
    
    df['Release_Date'][i]=data['datePublished']
        
    
    for val in details:
        v=val.text.split(':')[0]
        if len(re.compile('|'.join(val_required)).findall(v))>0:
            if v=="Language":
                df['Language'][i]= [x.text for x in val.findParent('div').find_all('a')]
            if v=="Budget":
                df['Budget'][i]= val.next_sibling.strip()
            if v=="Opening Weekend USA":
                df['opening_Weekend_USA'][i]= val.next_sibling.strip()
            if v=="Gross USA":
                df['Gross_USA'][i]= val.next_sibling.strip()
            if v=="Cumulative Worldwide Gross":
                df['Cumm_world_gross'][i]= val.next_sibling.strip()
            if v=="Runtime":   
                df['runtime'][i]=val.findParent('div').find('time').text.strip()    
                
                
#Self note :
#if used details for handling Director/Directors and writer/Writers we could use this:               
#         if len(re.compile('|'.join(val_required)).findall(v))>0:
#             if len(re.compile("Director").findall(v))>0:
#                 df['Directors'][i]= [x.text for x in val.findParent('div').find_all('a')]
#             if len(re.compile("Writer").findall(v))>0:
#                 df['Writers'][i]=[x.text for x in val.findParent('div').find_all('a')]            
```
